# Multi-Variant Analysis

## Zoning vs State vs Price

```python
top_states = df.groupby("state")["predicted_price_sqft_2026"].mean().nlargest(10).index

filtered_df = df[df["state"].isin(top_states)]

plt.figure(figsize=(10,5))

sns.barplot(
    x="state",
    y="predicted_price_sqft_2026",
    hue="zoning",
    data=filtered_df
)

plt.xticks(rotation=45)
plt.title("Top States vs Price (Zoning Wise)")
plt.show()
```

**Insight:**
- Compare land price across states
- Inside each state → compare zoning types
   - Example:
     - Commercial lands in some states are much higher
     - Residential may dominate in others
